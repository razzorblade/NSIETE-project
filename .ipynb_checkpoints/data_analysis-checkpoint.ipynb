# Data analysis

In order to be able to train a functional denoising model, we have to get a database of images and pre-process them in a way that allows us to run a neural network on it. 

We are using LSUN dataset that we have found in a project proposal. 

````python
import sys
import matplotlib.pyplot as plt
import numpy as np
import random, os
import cv2
from skimage.exposure import rescale_intensity
from skimage import data, color
from skimage.transform import rescale, resize, downscale_local_mean
from scipy import ndimage
from scipy import misc as sm
from os import listdir
from os.path import isfile, join

sys.path.append('..')
````

## Grayscale dataset
We want to test a noise removal from a grayscale images. In order to achieve this, we have to do some pre-processing.

We will use a ```church_dataset``` from the LSUN, which provides the largest number of images from the whole LSUN dataset. We will iterate through all files and load them. Then, we will rescale loaded images and save them in an output folder for the grayscale dataset.

````python
dataset_path = "./noise_generation/lsun/church_dataset"
````

````python
random.seed();

def recursive_files(dir):
    """Used to recursively iterate through all files in specified dir"""
    
    files = []
    for r, d, f in os.walk(dir):
        for file in f:
                files.append(os.path.join(r, file))
    return files

rfiles = recursive_files(dataset_path)

````

Let's print number of loaded files and example of what we have loaded:

````python
print(len(rfiles))
print("Loaded element 0: ", rfiles[0])
````

We have loaded 10,022 images which we can use to train our model. For this checkpoint, we will process only 1400. We will work with them in an implementation part.

````python
n=1400            #number of images to load
img_scale = 128   #scale of the images
grayscale_output = f"./noise_generation/grayscale/"      #where to save images

if not os.path.exists(grayscale_output):
    os.mknod(grayscale_output)
````

````python
for iteration in range(n):
    rfile = rfiles[iteration]
    
    img = cv2.imread(rfile)
    img = resize(img, (img_scale, img_scale), anti_aliasing=True)
    img = np.float32(img)

    gray = cv2.cvtColor(img*256, cv2.COLOR_BGR2GRAY)
    
    r_parts = rfile.split('/')
    r_end = r_parts[len(r_parts)-1]
    no_suffix = r_end.split('.')[0]
    
    cv2.imwrite(f"{grayscale_output}{no_suffix}.jpg", gray)
    print(f"[{iteration+1}/{n}] {rfile} -> ", f"{grayscale_output}{no_suffix}.jpg")
````

At this point, we have 1400 loaded grayscale images in ./noise_generation/grayscale folder.

````python
grayscale_paths = [f for f in listdir(grayscale_output)[:9] if isfile(join(grayscale_output, f))]
grayscale_loaded_imgs = np.empty([9,img_scale,img_scale,1])

#load images
for iteration,image_name in enumerate(grayscale_paths):
    name = grayscale_output+"/"+image_name
    read = cv2.imread(name,cv2.IMREAD_GRAYSCALE)
    grayscale_loaded_imgs[iteration] = read[:, :, np.newaxis]/255.0

    
_, axs = plt.subplots(3, 3, figsize=(6, 6))
axs = axs.flatten()
plt.gray()
for img, ax in zip(grayscale_loaded_imgs, axs):
    ax.imshow(img[:,:,0])
plt.show()
````

We have displayed 9 grayscale images from the output of our pre-processing. Let's print basic information about the first image. This will be common for all other images.

````python
print('Type of the image: ',type(grayscale_loaded_imgs))
print(f'Shape of the image: {grayscale_loaded_imgs[0].shape}')
print(f'Image Height: {grayscale_loaded_imgs[0].shape[0]}')
print(f'Image Width: {grayscale_loaded_imgs[0].shape[1]}')
print(f'Color channels of image {grayscale_loaded_imgs[0].shape[2]}')
print(f'Dimension of image: {grayscale_loaded_imgs[0].ndim}')
````

We can also see that with this load type, images have their pixels represented in 0..1 range, which is required by the neural network to work properly.

## RGB Dataset

There is not much work to do to get rgb dataset. We will just take images from LSUN and rescale them and then put them to their folder according to number of images to load.

````python
rgb_output = "./noise_generation/rgb/"

if not os.path.exists(rgb_output):
    os.mknod(rgb_output)
````

````python
random.seed();

rfiles = recursive_files(dataset_path)

for iteration in range(n):
    rfile = rfiles[iteration]
    
    img = cv2.imread(rfile, cv2.IMREAD_COLOR)
    img = resize(img, (img_scale, img_scale), anti_aliasing=True)

    r_parts = rfile.split('/')
    r_end = r_parts[len(r_parts)-1]
    no_suffix = r_end.split('.')[0]
    
    cv2.imwrite(f"{rgb_output}{no_suffix}.jpg", img*256)
    print(f"[{iteration+1}/{n}] {rfile} -> {rgb_output}{no_suffix}.jpg")
````

We have also implemented some functions that we can work later with. For example, we can generate salt and pepper noise or swap image pixels.

````python
def clamp(val, frm, to):
    if (val <= frm):
        return frm
    if (val >= to):
        return to
    return val

def apply_mat(image,i,j, mat_app):
    vr = vg = vb = 0
    clampi = i+len(mat_app)
    clampj = j+len(mat_app[0])
    
    if(clampi > len(image) or clampj > len(image[0])):
        return
    
    for c in range(i, clampi):
        for d in range(j, clampj):
            vr += image[c][d][0] * mat_app[c-i][d-j]
            vg += image[c][d][1] * mat_app[c-i][d-j]
            vb += image[c][d][2] * mat_app[c-i][d-j]
    image[i][j][0] = clamp(vr,0,255)        
    image[i][j][1] = clamp(vg,0,255)        
    image[i][j][2] = clamp(vb,0,255)
    
def apply_multichannel_mat(image, i, j, kernel_channels):
    vr = vg = vb = 0
    clampi = i+len(mat_app)
    clampj = j+len(mat_app[0])
    
    if(clampi > len(image) or clampj > len(image[0])):
        return
    
    for c in range(i, clampi):
        for d in range(j, clampj):
            vr += image[c][d][0] * kernel_channels[0][c-i][d-j]
            vg += image[c][d][1] * kernel_channels[1][c-i][d-j]
            vb += image[c][d][2] * kernel_channels[2][c-i][d-j]
    image[i][j][0] = clamp(vr,0,255)        
    image[i][j][1] = clamp(vg,0,255)        
    image[i][j][2] = clamp(vb,0,255)
    
    
def overlap(orig, other, other_opacity):
    output = np.copy(orig)
    for i in range(len(orig)):
        for j in range(len(orig[0])):
            output[i][j] = (1-other_opacity)*orig[i][j] + other_opacity*other[i][j]
    return output


def random_noise(orig, amount = 1, artifact_size = 0):
    output = np.copy(orig)
    for i in range(len(orig)):
        for j in range(len(orig[0])):
            if(np.random.rand() < amount):
                value = np.random.rand() *output[i][j] 
                output[i][j] = value
                for x in range(i-artifact_size,i + artifact_size):    
                    if (x > 0 and x < len(orig)):
                        output[x][j] = value
                        
                for y in range(j-artifact_size,j + artifact_size):    
                    if (y > 0 and y < len(orig[0])):
                        output[i][y] = value
    return output

def salt_and_pepper(orig, amount = 1):
    output = np.copy(orig)
    for i in range(len(orig)):
        for j in range(len(orig[0])):
            if(np.random.rand() < amount):
                if(np.random.rand() < 0.5):
                    output[i][j] = (0,0,0)
                else:
                    output[i][j] = (1,1,1)
    return output

def swap_pixels(orig, amount = 1):
    output = np.copy(orig)
    h = len(orig)
    w = len(orig[0])
    for i in range(len(orig)):
        for j in range(len(orig[0])):
            if(np.random.rand() < amount):
                defaultp = output[i][j]
                gi, gj = np.random.randint(0,h), np.random.randint(0,w)
                output[i][j] = output[gi][gj]
                output[gi][gj] = defaultp
                
    return output

def color_int(col):
    return max(min(round(col), 255), 0)

def boxed2flat(row):
    return [_color_int(x) for box in row for x in box]

def image2file(image, path):
    img = image
    with open(path, 'wb') as f:
        png.Writer(width=len(image[0]), height=len(image)).write(f, [boxed2flat(r) for r in img])
````
